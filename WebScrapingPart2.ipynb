# Amazon laptop search

```python
from selenium import webdriver
from bs4 import BeautifulSoup
```

```python
driver = webdriver.Chrome(executable_path='/Users/amritkaurchadha/Desktop/chromedriver')
driver.get('https://www.amazon.in/')
input_ele = driver.find_element_by_id('twotabsearchtextbox')
input_ele.send_keys('Laptops')
input_ele.submit()
data = driver.page_source
html_data = BeautifulSoup(data)
li = html_data.find_all('h2')
names = []
for i in li:
    if i.get('data-attribute') is not None:
        names.append(i.get('data-attribute'))

li2 = html_data.find_all(class_='a-price-whole')
prices = []
for i in li2:
    if i is not None:
        prices.append(i.string)
    
names
```

```python
import pandas as pd
df = pd.DataFrame({'Names':names,'Price':prices})
df
```
